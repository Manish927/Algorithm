```python
class Employee:
    total_employees = 0
    def __init__(self, name, department):
        self.name = name
        self.department = department
        Employee.total_employees += 1

    @classmethod
    def get_total_employees(cls):
        return cls.total_employees

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Department: {self.department}")



#create some employees
emp1 = Employee("Ayesha", "HR")
emp2 = Employee("Ravi", "Finance")
emp3 = Employee("Nina", "Engineering")

#Show details
emp1.display_info()
print()
emp2.display_info()
print()
emp3.display_info()
print()
print(f"Total employees: {Employee.get_total_employees()}")
```

```python
class Employee:
  def __init__(self, name, salary):
    self.name = name
    self.salary = salary

  def __str__(self):
    return f"Employee: {self.name}, Salary: {self.salary}"

  @staticmethod
  def calculate_tax(salary):
    return salary * 0.10

emp1 = Employee("Ayesha", 50000)
emp2 = Employee("Ravi", 60000)

print(emp1)
print(emp2)
tx = Employee.calculate_tax(emp1.salary)
print(f"Tax for {emp1.name}: {tx}")
tx = Employee.calculate_tax(emp2.salary)
print(f"Tax for {emp2.name}: {tx}")
```

```python
class Employee:
  def __init__(self, name, salary):
    self.name = name
    self.salary = salary

  def display_info(self):
    print(f"Name: {self.name}")
    print(f"Salary: {self.salary}")

  def __str__ (self):
    return f"Employee: {self.name}, Salary: {self.salary}"


class Manager(Employee):

  def __init__(self, name, salary, team_size):
    super().__init__(name, salary)
    self.team_size = team_size

  def display_info(self):
    super().display_info()
    print(f"Manages Team of: {self.team_size} people")


mgr = Manager("Ayesha", 50000, 5)
mgr.display_info()
print(mgr)
```

```python
from abc import ABC, abstractmethod

class TeamMember(ABC):
  @abstractmethod
  def role(self):
    pass

class Developer(TeamMember):
  def role(self):
    print("Developer")

class Tester(TeamMember):
  def role(self):
    print("Tester")

class Designer(TeamMember):
  def role(self):
    print("Designer")

members = [Developer(), Designer(), Tester()]

for member in members:
  member.role()
```
